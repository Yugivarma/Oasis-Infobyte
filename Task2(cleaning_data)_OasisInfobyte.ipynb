```python
# Data Wrangling

# Dropping id, host_name, last_review columns as they are unnecessary and will not impact our analysis and dataset
data.drop(['id','host_name','last_review'], axis=1, inplace=True)


# Filling null entries in name column with 'Unknown'
data['name'].fillna('Unknown', inplace=True)


# Filling null entries of reviews_per_month column with mean of all the other values of that column
mean_reviews_per_month = data['reviews_per_month'].mean()
data['reviews_per_month'].fillna(mean_reviews_per_month, inplace=True)


# Checking null entries after filling Null values
data.isnull().sum()
```

```python
# Visualising who are the top 10 hosts
top_hosts = data['host_id'].value_counts().head(10)
top_hosts_id = list(map(lambda x : str(x) , top_hosts.keys()))
top_hosts_count = list(top_hosts)

plt.figure(figsize=(10, 5))
plt.bar(top_hosts_id, top_hosts_count, color = 'r')
plt.grid(True)
plt.title('Top 10 Hosts')
plt.xticks(rotation = 30)
plt.xlabel('Top Hosts ID')
plt.ylabel('Number of times they hosted')
plt.show()
```

```python
# Setting the style for Seaborn
sns.set(style="whitegrid")

# Creating a countplot for the 'neighbourhood_group' column
plt.figure(figsize=(8, 4))
sns.countplot(x='neighbourhood_group', data=data, palette='viridis' )

# Customizing the plot
plt.title('Distribution of Listings Across Neighborhood Groups')
plt.xlabel('Neighborhood Group')
plt.ylabel('Number of Listings')
plt.show()
```

```python
# Distribution of room_type
plt.figure(figsize=(8, 4))
sns.countplot(x = 'room_type', data= data, palette='pastel')
plt.title('Distribution of Room Type')

# Customizing the plot
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing Distribution of Listing Prices
plt.figure(figsize=(8, 4))
sns.histplot(data['price'], bins=30, color='b')
plt.title('Distribution of Listing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualising distribution of prices for each neighberhood_group

# As we above see from data.describe statistical table that we have some extreme values, so i removed them for the sake of a better visualization.
df = data[data.price<1000]
plt.figure(figsize=(12,8))
sns.violinplot(x="neighbourhood_group", y="price", data=df, scale="width",palette="Set3")
plt.title('Density and distribution of prices for each neighbourhood_group')
plt.show()
```

```python
# Visualising distribution of reviews_per_month for each neighberhood_group

plt.figure(figsize = (8,4))
plt.title('Density and distribution of reviews_per_month for each neighbourhood_group')
sns.barplot(x='neighbourhood_group', y="reviews_per_month", data=data)
plt.show()
```
